```python
import pandas as pd
import os
from tqdm import tqdm
import matplotlib.pyplot as plt

# 이미지가 저장된 디렉토리
image_dir = './bank_statement'
# 처리할 이미지 확장자 목록
allowed_extensions = ['.png', '.jpg', '.jpeg']

# 디렉토리에서 이미지 파일 목록을 가져옵니다.
try:
    image_files = [f for f in os.listdir(image_dir) if os.path.splitext(f)[1].lower() in allowed_extensions]
except FileNotFoundError:
    print(f"오류: '{image_dir}' 폴더를 찾을 수 없습니다. 폴더를 생성하고 이미지를 넣어주세요.")
    image_files = []

# 모든 이미지의 처리 결과를 담을 리스트
all_processed_data = []

# 각 이미지 파일을 순회하며 OCR 수행
for filename in tqdm(image_files, desc="이미지 처리 중"):
    image_path = os.path.join(image_dir, filename)
    
    # --- 수정된 부분: debug=True로 변경하여 결과 이미지를 받음 ---
    # ocr.run_ocr은 debug=True일 때 (원본 이미지, ROI 이미지) 튜플을 반환합니다.
    result_images = ocr.run_ocr(image_path, debug=True)
    ocr_result = ocr.get_ocr_result()
    
    # --- 추가된 부분: 결과 이미지 출력 ---
    if result_images and len(result_images) == 2:
        # ROI(Region of Interest) 이미지를 출력합니다.
        roi_image = result_images[1] 
        plt.figure(figsize=(10, 12))
        plt.imshow(roi_image)
        plt.title(f"OCR Result for: {filename}")
        plt.axis('off') # 축 정보 숨기기
        plt.show()
    # --- 여기까지 추가 ---

    # OCR 결과가 있을 경우에만 데이터 처리
    if ocr_result and ocr_result.get('description'):
        for text, bbox_info in zip(ocr_result['description'], ocr_result['bounding_poly']):
            vertices = bbox_info['vertices']
            
            x_coords = [vertex['x'] for vertex in vertices]
            y_coords = [vertex['y'] for vertex in vertices]
            
            x_min, y_min = min(x_coords), min(y_coords)
            x_max, y_max = max(x_coords), max(y_coords)
            
            all_processed_data.append({
                'image_id': filename,
                'text': text,
                'x_min': x_min, 'y_min': y_min,
                'x_max': x_max, 'y_max': y_max,
                'label': 'O'
            })

# 모든 결과를 하나의 DataFrame으로 합칩니다.
final_df = pd.DataFrame(all_processed_data)

# DataFrame을 CSV 파일로 저장합니다.
output_csv_path = 'ocr_results2.csv'
final_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')

print(f"\n모든 이미지 처리가 완료되었습니다. 결과가 '{output_csv_path}' 파일에 저장되었습니다.")

# 저장된 결과의 일부를 확인합니다.
display(final_df.head())
```

```python
import pandas as pd
import re

def parse_amount(text, label):
    """Amount 텍스트에서 숫자만 추출하고, label에 따라 부호를 결정하는 함수"""
    if text is None:
        return None
    try:
        cleaned_text = re.sub(r'[^\d.-]', '', str(text))
        amount = float(cleaned_text)
        if label == 'AMOUNT_OUT' and amount > 0:
            return -amount
        return amount
    except (ValueError, TypeError):
        return None

# 1. 라벨링된 CSV 파일 로드
try:
    labeled_df = pd.read_csv('ocr_results.csv')
    print("ocr_results.csv 파일을 성공적으로 불러왔습니다.")
except FileNotFoundError:
    print("오류: ocr_results.csv 파일을 찾을 수 없습니다. 파일명을 확인해주세요.")
    labeled_df = None

if labeled_df is not None:
    # 최종 결과를 담을 리스트 초기화
    all_structured_transactions = []

    # 이미지 ID 별로 그룹화하여 처리
    for image_id, group in labeled_df.groupby('image_id'):
        
        # '사람이 보듯이' 위에서 아래로, 왼쪽에서 오른쪽으로 정렬
        group = group.sort_values(by=['y_min', 'x_min']).reset_index(drop=True)
        
        # --- ★★★ 최종 로직: 순차적 상태 기반 파서 ★★★ ---
        
        # 현재 이미지에서 처리 중인 정보를 담을 변수들
        current_transaction = {}
        last_known_date = None

        # 정렬된 순서대로 OCR 결과를 하나씩 순회
        for _, item in group.iterrows():
            label = item['label']
            text = item['text']

            # 1. 날짜 정보 업데이트
            if label == 'DATE':
                last_known_date = text
                # 만약 날짜가 거래처/금액보다 늦게 나와 이전 거래가 이미 완성된 경우,
                # 마지막 거래의 날짜를 업데이트 해줄 수 있음 (선택적)
                if all_structured_transactions and all_structured_transactions[-1]['image_id'] == image_id and all_structured_transactions[-1]['date'] is None:
                    all_structured_transactions[-1]['date'] = last_known_date


            # 2. 거래처 정보 수집
            elif label == 'MERCHANT':
                # 만약 이미 '현재 거래'에 거래처가 있는데 새로운 거래처가 나오면,
                # 이전 거래는 완성되지 못했으므로 버리고 새로 시작.
                if 'merchant' in current_transaction:
                    current_transaction = {} # 초기화
                current_transaction['merchant'] = text

            # 3. 금액 정보 수집
            elif label in ['AMOUNT_IN', 'AMOUNT_OUT']:
                if 'amount' in current_transaction:
                    current_transaction = {} # 이미 금액이 있는데 또 나오면 초기화
                current_transaction['amount'] = parse_amount(text, label)

            # 4. 거래 완성 조건 확인
            # '거래처'와 '금액'이 모두 채워졌다면 하나의 거래로 완성!
            if 'merchant' in current_transaction and 'amount' in current_transaction:
                # 완성된 거래에 이미지 ID와 날짜를 추가
                current_transaction['image_id'] = image_id
                current_transaction['date'] = last_known_date
                
                # 최종 결과 리스트에 추가
                all_structured_transactions.append(current_transaction)
                
                # '현재 거래'를 초기화하여 다음 거래를 준비
                current_transaction = {}

    # 최종 결과 DataFrame 생성
    final_df = pd.DataFrame(all_structured_transactions)

    # 결과 저장 및 출력
    output_path = 'transactions_processed_essentials.csv'
    final_df.to_csv(output_path, index=False, encoding='utf-8-sig')
    
    print(f"\n핵심 정보 추출이 완료되었습니다. 결과가 '{output_path}' 파일에 저장되었습니다.")
    display(final_df)
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta
import os

# --- 설정값 ---
# train_bert_all_cnn4.py에서 사용하는 거래처 데이터 파일 경로
MERCHANT_DATA_PATH = './processed_data/region_all_processed_data.csv'

# --- 데이터 로딩 ---
# 거래처명 데이터 로드 (파일이 없을 경우 예외 처리)
try:
    merchant_df = pd.read_csv(MERCHANT_DATA_PATH)
    # '상호명_Regulated' 컬럼의 값을 리스트로 변환, 중복 제거
    merchant_list = merchant_df['상호명_Regulated'].dropna().unique().tolist()
    print(f"'{MERCHANT_DATA_PATH}'에서 {len(merchant_list)}개의 고유 거래처명을 불러왔습니다.")
except FileNotFoundError:
    print(f"경고: 거래처 데이터 파일('{MERCHANT_DATA_PATH}')을 찾을 수 없습니다. 가상 거래처명을 사용합니다.")
    # Faker를 대체할 간단한 가상 거래처명 생성
    merchant_list = [f'가상상점_{i}' for i in range(200)]
    merchant_list.extend([f'가상회사_{i}(주)' for i in range(50)])


def generate_realistic_transactions(num_transactions, merchants):
    """
    사용자 요구사항에 맞춰 현실성을 높인 가상 거래내역 데이터프레임을 생성합니다.
    """
    if not merchants:
        print("오류: 거래처 목록이 비어있습니다. 거래 데이터를 생성할 수 없습니다.")
        return pd.DataFrame()

    transactions = []
    
    # 1. 날짜는 오늘부터 한달 이내 랜덤한 시점으로 설정
    # 가장 최근 거래 날짜를 오늘 ~ 30일 전 사이에서 랜덤하게 설정
    current_date = datetime.now() - timedelta(days=random.randint(0, 29))

    for _ in range(num_transactions):
        
        # 4. 거래 날짜는 일정한 확률로 이동
        # 40% 확률로 같은 날, 30% 확률로 하루 전, 20% 확률로 이틀 전, 10% 확률로 사흘 전
        rand_val = random.random()
        if rand_val < 0.4:
            day_delta = 0
        elif rand_val < 0.7:
            day_delta = 1
        elif rand_val < 0.9:
            day_delta = 2
        else:
            day_delta = 3
        
        current_date -= timedelta(days=day_delta)
        # 같은 날이라도 다른 거래처럼 보이도록 시간은 랜덤하게 설정
        transaction_date = current_date.replace(hour=random.randint(0,23), minute=random.randint(0,59), second=random.randint(0,59))

        # 2. 실제 거래처명 중 랜덤으로 하나 선택
        merchant_name = random.choice(merchants)
        
        # 3. 출금/입금 비율 및 금액 조정
        if random.random() < 0.9: # 90% 확률로 출금
            amount = -random.randint(10, 2000) * 100
        else: # 10% 확률로 입금 (월급 등 고려)
            amount = random.randint(50, 40000) * 100
            
        transactions.append({
            'date': transaction_date.strftime('%Y.%m.%d'), # 날짜 형식은 기존과 동일하게 유지
            'merchant': merchant_name,
            'amount': float(amount)
        })
        
    return pd.DataFrame(transactions)

# 1000 ~ 10000건 사이의 가상 거래내역 생성
num_to_generate = random.randint(1000, 10000)
synthetic_df = generate_realistic_transactions(num_to_generate, merchant_list)

print(f"\n현실성을 반영한 가상 거래 데이터 {len(synthetic_df)}건 생성이 완료되었습니다.")
display(synthetic_df)
```

```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import os
from tqdm import tqdm
import pandas as pd
import random
from datetime import datetime, timedelta
import glob

# --- 1. 설정 (너비 임계값 추가) ---
COLOR_TO_LABEL_MAP = {
    (255, 0, 255): 'DATE',
    (0, 255, 255): 'DATE_HEADER',
    (0, 255, 0):   'MERCHANT',
    (0, 0, 255):   'MEMO',
    (255, 0, 0):   'AMOUNT',
    (255, 255, 0): 'BALANCE',
    (255, 127, 255): 'TIME',
}
output_dir = './generated_bank_statements_final'
os.makedirs(output_dir, exist_ok=True)
label_output_file = os.path.join(output_dir, 'generated_labels.csv')
MULTILINE_HEIGHT_THRESHOLD = 55
DATE_WIDTH_THRESHOLD = 130 # YYYY.MM.DD 형식을 제외할 DATE 박스의 최소 너비

# --- 2. 폰트 로딩 (변경 없음) ---
font_dir = './fonts'
font_sizes_to_load = [32, 36, 40, 44] 
loaded_fonts = []
try:
    font_files = [f for f in os.listdir(font_dir) if f.lower().endswith('.ttf')]
    if not font_files: raise FileNotFoundError("폰트 파일을 찾을 수 없습니다.")
    for font_file in font_files:
        for size in font_sizes_to_load:
            loaded_fonts.append(ImageFont.truetype(os.path.join(font_dir, font_file), size))
    print(f"'{font_dir}'에서 {len(font_files)}개 폰트 파일을 로드하여 {len(loaded_fonts)}개의 폰트 객체를 생성했습니다.")
except FileNotFoundError as e:
    print(f"경고: '{font_dir}' 폴더에 .ttf 폰트 파일이 없습니다. ({e})")
    loaded_fonts = [ImageFont.load_default()]

# --- 3. 템플릿 파싱 (변경 없음) ---
def parse_layout_from_color_template(template_path, color_map):
    try:
        image = cv2.imread(template_path)
        if image is None: return None
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    except Exception: return None
    layout = {label: [] for label in color_map.values()}
    for color, label in color_map.items():
        mask = cv2.inRange(image_rgb, np.array(color), np.array(color))
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            layout[label].append((x, y, x + w, y + h))
    for label in layout:
        layout[label].sort(key=lambda bbox: (bbox[1], bbox[0]))
    return layout

# --- 4. 데이터 생성 (날짜 형식 인자 추가) ---
def generate_transaction_data(num_items, merchants, config, allowed_date_formats):
    drawable_items = []
    memos = ['#체크카드', '#용돈', '#월급', '#이체', '#송금', '#카드결제', '#오픈뱅킹이체']
    current_date = datetime.now() - timedelta(days=random.randint(0, 30))
    balance = random.randint(500000, 2000000)
    
    last_item_date_str = ""
    date_style = random.choice(config['date_style_options'])
    chosen_date_format = random.choice(allowed_date_formats)
    transaction_dates = []

    for i in range(num_items):
        day_delta = random.choices([0, 1, 2, 3], weights=[0.4, 0.3, 0.2, 0.1])[0]
        current_date -= timedelta(days=day_delta, hours=random.randint(1, 5))
        transaction_dates.append(current_date)
        date_str, time_str = chosen_date_format(current_date), current_date.strftime('%H:%M')

        if date_style == 'per_item' or (date_style == 'toss_like' and date_str != last_item_date_str):
            drawable_items.append({'label': 'DATE', 'text': date_str, 'item_index': i})
        last_item_date_str = date_str
        drawable_items.append({'label': 'TIME', 'text': time_str, 'item_index': i})

        merchant_name = random.choice(merchants)
        is_income = random.random() < 0.15
        amount_val = random.randint(500, 4000) * 100 if is_income else random.randint(10, 200) * 100
        
        if config['amount_style'] == 'text':
            label, text = ('AMOUNT_IN', f"입금 {amount_val:,}원") if is_income else ('AMOUNT_OUT', f"출금 {amount_val:,}원")
        else:
            amount_val = amount_val if is_income else -amount_val
            label, text = ('AMOUNT_IN', f"{amount_val:,}원") if is_income else ('AMOUNT_OUT', f"{amount_val:,}원")
        
        balance += amount_val
        drawable_items.extend([
            {'label': 'MERCHANT', 'text': merchant_name, 'item_index': i},
            {'label': label, 'text': text, 'item_index': i},
            {'label': 'BALANCE', 'text': f"{balance:,}원", 'item_index': i}
        ])
        if random.random() < 0.7:
            drawable_items.append({'label': 'MEMO', 'text': random.choice(memos), 'item_index': i})

    if transaction_dates and parsed_layout.get('DATE_HEADER'):
        start_date, end_date = min(transaction_dates), max(transaction_dates)
        header_format_options = [
            lambda s, e, n: f"{e.strftime('%Y.%m.%d')} ~ {s.strftime('%Y.%m.%d')} ({n}건)",
            lambda s, e, n: f"{e.strftime('%Y년 %m월')}"
        ]
        header_text = random.choice(header_format_options)(start_date, end_date, num_items)
        drawable_items.append({'label': 'DATE_HEADER', 'text': header_text, 'item_index': 0})
    return drawable_items

# --- 5. 이미지 생성 (날짜 형식 동적 선택 로직 추가) ---
def generate_synthetic_images(template_configs, num_images=20):
    if not loaded_fonts or not template_configs: return
    all_labels_df = pd.DataFrame()
    amount_color_schemes = {
        'blue_red':   {'AMOUNT_IN': (50, 100, 255), 'AMOUNT_OUT': (200, 30, 30)},
        'blue_black': {'AMOUNT_IN': (50, 100, 255), 'AMOUNT_OUT': (20, 20, 20)}
    }
    print(f"\n총 {num_images}개의 다양한 은행 명세서 이미지 생성을 시작합니다...")

    for i in tqdm(range(num_images), desc="이미지 생성 중"):
        chosen_config = random.choice(template_configs)
        global parsed_layout 
        parsed_layout = parse_layout_from_color_template(chosen_config['colored_path'], COLOR_TO_LABEL_MAP)
        if not parsed_layout or not parsed_layout.get('MERCHANT'): continue
            
        # ★★★ DATE 박스 너비에 따라 날짜 형식 동적 선택 ★★★
        allowed_date_formats = [
            lambda d: d.strftime('%Y.%m.%d'),
            lambda d: d.strftime('%m.%d'),
            lambda d: f"{d.month}월 {d.day}일"
        ]
        if 'DATE' in parsed_layout and parsed_layout['DATE']:
            date_box_width = parsed_layout['DATE'][0][2] - parsed_layout['DATE'][0][0]
            if date_box_width < DATE_WIDTH_THRESHOLD:
                allowed_date_formats = [
                    lambda d: d.strftime('%m.%d'),
                    lambda d: f"{d.month}월 {d.day}일"
                ]

        items_per_image = len(parsed_layout.get('MERCHANT', []))
        img = Image.open(chosen_config['clean_path']).convert("RGB")
        draw = ImageDraw.Draw(img)
        font = random.choice(loaded_fonts)
        image_data = generate_transaction_data(items_per_image, merchant_list, chosen_config, allowed_date_formats)
        image_labels = []
        chosen_color_scheme = random.choice(list(amount_color_schemes.values()))

        for item in image_data:
            label, text, item_idx = item['label'], item['text'], item['item_index']
            layout_label = 'AMOUNT' if 'AMOUNT' in label else label
            current_item_idx = 0 if label == 'DATE_HEADER' else item_idx

            if layout_label in parsed_layout and current_item_idx < len(parsed_layout[layout_label]):
                x_min, y_min, x_max, y_max = parsed_layout[layout_label][current_item_idx]
                
                is_multiline = False
                if label in ['AMOUNT_IN', 'AMOUNT_OUT'] and chosen_config['amount_style'] == 'text':
                    height = y_max - y_min
                    if height > MULTILINE_HEIGHT_THRESHOLD:
                        text = text.replace(' ', '\n', 1)
                        is_multiline = True

                align_right = label in ['AMOUNT_OUT', 'AMOUNT_IN', 'BALANCE', 'TIME']
                draw_x = x_max - draw.textlength(text.split('\n')[-1], font=font) if align_right else x_min
                
                if is_multiline:
                    draw.multiline_text((draw_x, y_min), text, font=font, fill=chosen_color_scheme.get(label, (20,20,20)), align="right" if align_right else "left")
                    final_bbox = draw.multiline_textbbox((draw_x, y_min), text, font=font)
                else:
                    text_bbox = draw.textbbox((0,0), text, font=font)
                    draw_y = y_min + (y_max - y_min - (text_bbox[3] - text_bbox[1])) / 2
                    color_map = {'MEMO': (0, 150, 200), 'DATE_HEADER': (100, 100, 100), 'DATE': (100, 100, 100), 'TIME': (120, 120, 120), 'BALANCE': (150, 150, 150)}
                    color = chosen_color_scheme.get(label, color_map.get(label, (20, 20, 20)))
                    draw.text((draw_x, draw_y), text, font=font, fill=color)
                    final_bbox = draw.textbbox((draw_x, draw_y), text, font=font)

                image_labels.append({'image_id': f'final_gen_{i+1}.png', 'text': text.replace('\n', ' '), 'x_min': final_bbox[0], 'y_min': final_bbox[1], 'x_max': final_bbox[2], 'y_max': final_bbox[3], 'label': label})

        img.save(os.path.join(output_dir, f'final_gen_{i+1}.png'), "PNG")
        if image_labels:
            all_labels_df = pd.concat([all_labels_df, pd.DataFrame(image_labels)], ignore_index=True)

    all_labels_df.to_csv(label_output_file, index=False, encoding='utf-8-sig')
    print(f"\n이미지 및 라벨 생성이 완료되었습니다. '{output_dir}' 폴더를 확인하세요.")
    display(all_labels_df.head())

# --- 6. 실행 ---
template_dir = './bank_templates'
template_files = glob.glob(os.path.join(template_dir, '*_clean.png'))
TEMPLATES_CONFIG = []
for clean_path in template_files:
    colored_path = clean_path.replace('_clean.png', '_colored.png')
    if os.path.exists(colored_path):
        name = os.path.basename(clean_path).replace('_clean.png', '')
        amount_style = 'text' if 'shinhan' in name.lower() or 'kookmin' in name.lower() else 'sign'
        date_style_options = ['per_item', 'toss_like'] if 'kakao' in name.lower() else ['per_item']
        TEMPLATES_CONFIG.append({'name': name, 'clean_path': clean_path, 'colored_path': colored_path, 'amount_style': amount_style, 'date_style_options': date_style_options})
        print(f"'{name}' 템플릿 설정 로드 완료.")

if 'merchant_list' in locals() and TEMPLATES_CONFIG:
    generate_synthetic_images(TEMPLATES_CONFIG, num_images=20)
elif not 'merchant_list' in locals():
    print("오류: 'merchant_list'가 정의되지 않았습니다. 이전 셀을 먼저 실행해주세요.")
else:
    print(f"오류: '{template_dir}' 폴더에서 유효한 템플릿 설정을 찾지 못했습니다.")
```
